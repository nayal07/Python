```python
df_flipkart['product_rating']= df_flipkart['product_rating'].astype(int)
```

```python
def remove_char (val):
    return(int(val.replace('₹', '').replace(',','')))
```

```python
df_flipkart.mrp = df_flipkart.mrp.map(remove_char)
```

```python
df_flipkart.mrp
```

```python
df_flipkart.selling_price = df_flipkart.selling_price.map(remove_char)
```

```python
sns.boxplot(x='category_1', y='product_rating', data = df_flipkart)
plt.title('Category_1 VS Product Rating\n')

plt.figure(figsize=(10,10))
plt.xticks(rotation=45)
plt.show()
```

```python
product_selling_price = df_flipkart.groupby(['category_3'])['selling_price'].agg(max).sort_values(ascending=False)[:10]
```

```python
product_selling_price
```

```python
plt.figure(figsize=(15, 8))

product_selling_price = df_flipkart.groupby(['category_3'])['selling_price'].agg(max).sort_values(ascending=False)[:10]

g= sns.barplot(x=product_selling_price.index, y=product_selling_price.values , palette='magma')

for i in range(10):
    value= product_selling_price[i]
    
    g.text(y=value -2 , x=i+0.125 , s=value, color='black', ha='center', fontsize=15)
    
    g.set_xlabel('product', fontsize=15)
    g.set_ylabel('seling_price', fontsize= 15)
    g.set_title('Top 10 products with heightest selling price\n',color = 'darkred', fontsize =  30)
    plt.xticks(rotation = 45)
    plt.show()
```

```python
vc=df_flipkart.groupby(['category_3'])['mrp'].agg(max).sort_values(ascending=False)[:10]

plt.figure(figsize=(15,10))
G= sns.barplot(x=vc.index, y=vc.values, data=df_flipkart, palette='Set2')

for i in range(10):
    value = vc[i]
    G.text(y=value -2, x=i +0.125, s= value , color = 'black',ha='center', fontsize=12)

plt.xticks(rotation=45)
plt.xlabel('Products',fontsize=12)
plt.ylabel('MRP', fontsize=15)
plt.title('Top 10 Products with Heighest MRP')    
plt.show()
```
